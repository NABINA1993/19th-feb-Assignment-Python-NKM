Q1. What is Flask Framework? What are the advantages of Flask Framework?

Ans:-Flask is a micro web framework for Python that allows developers to easily build web applications. It is designed to be lightweight, modular, and easy to use, while still providing the flexibility and power necessary for complex web applications. Some advantages of Flask framework are:

Lightweight: Flask is designed to be lightweight and simple, which means that it doesn't have a lot of unnecessary overhead. This makes it fast and efficient, which is important for web applications that need to be responsive and scalable.

Modular: Flask is built on a modular architecture, which means that it can be easily extended and customized. This makes it flexible and adaptable to a wide range of use cases.

Easy to learn: Flask has a simple and intuitive API that makes it easy for developers to learn and use. It also has a large and active community of developers, which means that there are many resources available for learning and troubleshooting.

Extensive documentation: Flask has extensive documentation that is well organized and easy to navigate. This makes it easy for developers to find the information they need and quickly get up and running with the framework.

Large ecosystem: Flask has a large ecosystem of plugins and extensions that can be used to extend its functionality. This makes it easy to add features like authentication, database integration, and more to your web application.

Overall, Flask is a powerful and flexible web framework that is well-suited to a wide range of web application development needs. Its lightweight, modular design and easy-to-learn API make it a great choice for developers who want to build robust, scalable web applications quickly and efficiently.

Q3. What is App routing in Flask? Why do we use app routes?

Ans:-In Flask, app routing refers to the process of mapping a URL to a view function, which generates the HTTP response for that URL. App routing is an essential part of building a Flask application, as it allows you to define the structure and behavior of your web application.

In Flask, app routes are defined using the @app.route() decorator. The decorator is used to associate a URL endpoint with a Python function that will handle requests to that endpoint. The decorator takes a string argument that represents the URL endpoint, and it can be followed by one or more HTTP methods that the endpoint will respond to (e.g. GET, POST, PUT, DELETE, etc.). The function that is decorated with @app.route() should return a string that represents the response to the client's request.

Here is an example of defining an app route in Flask:

python
Copy code
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'
In this example, we define an app route for the root URL (/). When the user visits the root URL, the index() function will be called, which simply returns the string "Hello, World!".

We use app routes in Flask to define the structure and behavior of our web application. App routes allow us to define the URLs that our users can visit, and they allow us to specify what happens when a user visits each URL. By defining app routes, we can create a complete web application that is capable of handling a wide range of user interactions.

Overall, app routing is an important concept in Flask and web development in general, as it provides a way to organize the behavior of our web application and respond to user requests in a structured and organized way.

Q4.  Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” 
route to show the following details:

    Company Name: ABC Corporation

    Location: India

    Contact Detail: 999-999-9999 
   
 Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def index():
    company_name = 'ABC Corporation'
    
    location = 'India'
    
    contact_detail = '999-999-9999'
    
    return f"Company Name: {company_name}\nLocation: {location}\nContact Detail: {contact_detail}"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the 
url_for() function.

Ans:-The url_for() function in Flask is used for URL building. It takes the name of the view function and generates a URL for that function. The advantage of using url_for() instead of hard-coding the URL in your templates or code is that if the URL of a view function changes, you only need to update it in one place - the function definition. Flask's url_for() function will automatically update all references to the function with the new URL.

Here is an example of how to use the url_for() function in Flask:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return '<a href="' + url_for('about') + '">About Us</a>'

@app.route('/about')
def about():
    return 'This is the About Us page.'

if __name__ == '__main__':
    app.run(debug=True)
```
